# 1. Selenium을 활용하여 크롤링하기

## 목표 : 멜론 사이트 멜론차트 TOP100 결과를 크롤링하여 데이터프레임으로 저장해보자!

```python
from selenium import webdriver
import time
import pandas as pd
```

### 1) 멜론 사이트에 접속하여 검색어 입력하고 클릭하기

```python
#자신의 chromedriver가 있는 파일 위치 경로 기억하기
#window는 ~/chromedriver.exe를 불러오고 mac은 ~/chromedriver를 불러오기
path = "C:/Users/Downloads/chromedriver_win32/chromedriver.exe"
```

```python
#selenium으로 제어할 수 있는 크롬 브라우저 창이 뜬다.
driver = webdriver.Chrome(path)
```

```python
#웹 페이지 접속하기
url = 'https://www.melon.com/'
#webdriver가 멜론사이트에 접속하도록 명령
driver.get(url)
```

```python
#검색창 element 찾기
#현재 열려 있는 창의 element의 id가 'top_search'인 것을 찾아서 element 변수에 할당
element = driver.find_element_by_id('top_search')
```

```python
#검색어 입력하기
element.clear()
element.send_keys('방탄소년단')
```

```python
#검색 클릭하기
finder = driver.find_element_by_css_selector('.search_m')
finder.click()
```

```python
#검색 클릭하기
finder = driver.find_element_by_css_selector('.menu01.mlog')
finder.click()
```

### 2) TOP50 곡들의 곡명, 아티스트 이름을 크롤링해서 프린트하기

```python
for i in range(50):
    print('곡명 : ', driver.find_elements_by_css_selector('.lst50 .wrap_song_info .rank01')[i].text)
    print('아티스트 : ', driver.find_elements_by_css_selector('.lst50 .wrap_song_info .rank02')[i].text)
```

### 3) TOP100곡들의 상세정보 페이지 접속에 필요한 노래 아이디 가져오기

```python
#TOP100 곡들의 song id 가져오기
song_ids_50 = driver.find_elements_by_css_selector('tr#lst50.lst50')
song_ids_50 = [song_id.get_attribute('data-song-no') for song_id in song_ids_50]
song_ids_100 = driver.find_elements_by_css_selector('tr#lst100.lst100')
song_ids_100 = [song_id.get_attribute('data-song-no') for song_id in song_ids_100]
song_ids = song_ids_50 + song_ids_100
print(song_ids)
```

### 4) 곡명, 아티스트, 앨범명, 발매일, 장르, 좋아요 수, 리뷰 수를 크롤링하여 프린트하기

```python
#1위곡 상세정보 페이지 접속
url = 'https://www.melon.com/song/detail.htm?songId=' + song_ids[0]
driver.get(url)
```

```python
print('곡명 : ' ,driver.find_element_by_css_selector('.song_name').text)
print('아티스트 : ',driver.find_element_by_css_selector('.artist').text)

for i in range(3):
    if i == 0:
        print('앨범명 : ',driver.find_elements_by_css_selector('.list dd')[i].text)
    elif i ==1:
        print('발매일 : ',driver.find_elements_by_css_selector('.list dd')[i].text)
    else:
        print('장르 : ',driver.find_elements_by_css_selector('.list dd')[i].text)
        
print('좋아요 갯수 : ', driver.find_element_by_css_selector('#d_like_count').text)
print('리뷰 갯수 : ',driver.find_element_by_css_selector('.d_cmtpgn_srch_cnt').text)
```

### 5) 멜론차트 TOP100 곡 상세정보를 데이터프레임에 저장하기

#### 주의! Stale Element Reference Exception 에러
- 내용
    - 1. The element has been deleted entirely.
    - 2. The element is no longer attached to the DOM.
- 해결
    - 1. time.sleep을 통한 명시적 대기
    - 2. driver.implicitly_wait

```python
result = list()

for i in range(len(song_ids)):
    
    #명시적 대기
    time.sleep(3)
    
    #곡 상세정보 페이지 접속
    url = 'https://www.melon.com/song/detail.htm?songId=' + song_ids[i]
    driver.get(url)
    
    #곡 정보 가져오기
    song_name = driver.find_element_by_css_selector('.song_name').text  #곡이름
    artist = driver.find_element_by_css_selector('.artist').text  #아티스트 이름
    album_name =  driver.find_elements_by_css_selector('.list dd')[0].text  #앨범이름
    date = driver.find_elements_by_css_selector('.list dd')[1].text  #발매일
    genre = driver.find_elements_by_css_selector('.list dd')[2].text  #장르
    like = driver.find_element_by_css_selector('#d_like_count').text  #좋아요 수
    review = driver.find_element_by_css_selector('.d_cmtpgn_srch_cnt').text  #리뷰 수
    
    #곡 정보 추가하기
    result.append({
                'song_name': song_name,
                'artist': artist,
                'album_name': album_name,
                'date': date,
                'genre': genre,
                'like': like,
                'review': review
                })
     
    print(i+1,'번째 곡 완료')
```

```python
#데이터프레임으로 만들기
result = pd.DataFrame(result, columns = ['song_name', 'artist', 'album_name', 'date', 'genre','like', 'review'])
result
```

```python
#webdriver를 종료하여 창닫기
driver.close()
```

# 2. API를 활용하여 크롤링하기

## 목표 : 카카오 API를 통해 원하는 정보를 크롤링하여 데이터프레임으로 저장해보자!

```python
import json, requests
```

### 1) 강남구 삼성동 반경 5km이내 올리브영 매장정보 읽어오기

```python
#API의 url 구성
#json : 키-값 쌍으로 이루어진 데이터 포맷
url = "https://dapi.kakao.com/v2/local/search/keyword.json"
#API KEY 지정
apikey = "자신의 rest api 키" 
#파라미터 지정
query = "올리브영" #검색을 원하는 질의어
x = 37.4834308 #강남구 삼성동 경도
y = 127.0616783 #강남구 삼성동 위도
radius = 5000 #반경 5km의 매장정보 읽어오기(단위: 미터)


#page를 설정하여 해당 페이지를 읽어올수 있음(최대 45페이지), 한페이지에 보여질 문서개수(size)는 최대 15개
#앱키를 헤더에 담아 API에 요청을 보내 데이터 추출
r = requests.get( url, params = {'query':query, 'x':x, 'y':y, 'radius':radius}, headers={'Authorization' : 'KakaoAK ' + apikey } ) 
#결과를 json 형식으로 반환
data = json.loads(r.text)
data
```

### 2) 올리브영 매장 상세정보를 데이터프레임에 저장하기

```python
#크롤링한 정보 데이터프레임으로 만들기
result2 = list()

for i in range(len(data['documents'])):
    place_name = data['documents'][i]['place_name'] # 매장이름
    address_name = data['documents'][i]['address_name'] #매장주소
    road_address_name = data['documents'][i]['road_address_name'] #매장 도로명주소
    tel = data['documents'][i]['phone'] #매장 전화번호
    place_url = data['documents'][i]['place_url'] #장소 상세페이지 url
    
    result2.append({
                'place_name': place_name,
                'address_name': address_name,
                'road_address_name': road_address_name,
                'tel': tel,
                'place_url': place_url
                })
```

```python
result2 = pd.DataFrame(result2, columns = ['place_name','address_name','road_address_name','tel','place_url'])
result2
```
